## Load image and plot

```python
# Load image
img_dir = Path('example-imgs')
if not img_dir.exists():
    print('Directory not found')
img_paths = [path for path in img_dir.glob('*.tif')]
img = iio.imread(img_paths[0])
# Plot image and intenisty
ums_per_pixel = 2  # Dummy value
fig, axes = lineprofile.plot_img_and_line(img, (250, 250), (500, 500), ums_per_pixel)
plt.show()
```
